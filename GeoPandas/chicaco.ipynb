```python
gdf3395 = gdf.to_crs("EPSG:3395")
```

```python
gdf
```

```python
gdf['area'] = gdf.area
```

```python
gdf3395
```

```python
gdf3395['area'] = gdf3395.area
```

```python
gdf3395.plot("area", legend=True)
```

```python
gdf3395['centroid'] = gdf3395.centroid
```

```python
gdf3395 = gdf3395.set_geometry('centroid')
gdf3395.plot("area", legend=True)
```

```python
gdf3395 = gdf3395.set_geometry('geometry')
ax = gdf3395['geometry'].plot()
gdf3395['centroid'].plot(ax = ax, color="orange")
```

```python
gdf3395['convex_hull'] = gdf3395.convex_hull
gdf3395['boundary'] = gdf3395.boundary
ax = gdf3395['convex_hull'].plot(alpha=.5)
gdf3395['boundary'].plot(ax=ax, color="white", linewidth=.3)
```

```python
gdf3395['buffered'] = gdf3395.buffer(500)
gdf3395['buffered_centroid'] = gdf3395['centroid'].buffer(700)
ax = gdf3395['buffered'].plot(alpha=.5)
gdf3395['buffered_centroid'].plot(ax=ax, color="red", alpha=.3)
gdf3395['boundary'].plot(ax=ax, color="white", linewidth=.3)
```
