```python
import pandas as pd

df = pd.read_csv('https://data.cityofchicago.org/resource/e55j-2ewb.csv')
```

```python
df.head()
```

```python
geometry = [Point(xy) for xy in zip(df['pickup_centroid_longitude'],df['pickup_centroid_latitude'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry, crs=4326)
```

```python
gdf
```

```python
gdf.plot()
```

```python
gdf = gdf.to_crs("EPSG:3395")
```

```python
gdf
```

```python
gdf['area'] = gdf.area
```

```python
gdf.plot("area", legend=True)
```
